```python
horarios_mañana = list(range(19, -1, -1))
print(horarios_mañana)
horarios_tarde = list(range(29, 19, -1))
print(horarios_tarde)
```

```python
def no_overlap(a, b):
    """ Verifica que dos clases no se solapen en horario """
    return a != b

def generate_schedule(df, horarios):
    """
    Genera un horario para las clases especificadas en el DataFrame df,
    utilizando la lista de horarios proporcionada.
    
    Args:
        df: DataFrame que contiene las clases y sus codificaciones.
        horarios: Lista de horarios disponibles.
    
    Returns:
        Diccionario de solución que asigna horarios a cada clase.
    """
    # Crear el problema CSP
    problem = Problem()
    
    # Añadir las variables (clases) y sus dominios (horarios)
    for clase in df['codificacion']:
        problem.addVariable(clase, horarios)
    
    # Añadir la restricción de no solapamiento
    for i in range(len(df)):
        for j in range(i + 1, len(df)):
            problem.addConstraint(no_overlap, (df['codificacion'][i], df['codificacion'][j]))
    
    # Obtener la solución
    solucion = problem.getSolution()
    solucion = {clase: dict_horario[horario] for clase, horario in solucion.items()}
    df['horario'] = df['codificacion'].apply(lambda x: solucion[x])
    return df
```
